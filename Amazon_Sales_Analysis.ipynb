# The "M" size is the most commonly purchased by customers.

```python
#courier status
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Courier Status', hue= 'Status')
plt.show()
```

# The majority of orders have a status of Shipped - Delivered to Buyers

```python
df['Category'] = df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,6))
plt.hist(column_data, bins=15, edgecolor="Black", color="Yellow")
plt.xticks(rotation=90)
plt.show()
```

# T-shirts account for the highest number of orders.

```python
#check B2B data 
check_B2B = df['B2B'].value_counts()

plt.pie(check_B2B, labels= check_B2B.index, autopct = '%1.1f%%')
plt.show()
```

# A significant 99.2% of buyers are retailers, while the remaining are wholesalers.

```python
#scatter plot
x_data= df['Category']
y_data=df['Size']

plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Available Size')
plt.show()
```

```python
#Top Ten States
top10_state = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,8))
sns.countplot(data=df[df['ship-state'].isin(top10_state.index)],x='ship-state')
plt.xlabel('State')
plt.ylabel('Order_count')
plt.title('Distribution of States')
plt.xticks(rotation =90)
plt.show()
```

# Maharashtra ranks at the top among the top 10 states with the highest number of orders.
